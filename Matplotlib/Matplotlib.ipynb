Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.


**Some of the major Pros of Matplotlib are:**

  Generally easy to get started for simple plots

  Support for custom labels and texts

  Great control of every element in a figure

  High-quality output in many formats

  Very customizable in general

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
import numpy as np
x=np.arange(0,10)
y=np.arange(10,20)
a=np.arange(40,50)
b=np.arange(50,60)
```

### **Scatter Plot**

```
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('Simple scatter plot')
plt.savefig('test.png')
```

```
y=x*x
```

```
plt.plot(x,y,'ro--',linewidth=1)
plt.xlabel('X axis')
plt.ylabel('Y label')
plt.title('Simple scatter plot')
```

### **Sub Plot**

```
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```
#one good looking example
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sin theta form')

#plot them
plt.plot(x,y)
plt.show()
```

```
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 2, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 2, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine') 
```

### **Bar Plot**

```
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,color='y')
plt.bar(x2,y2)
```

### **Histogram**

```
x=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
y=[22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]
print(x)
print(y)
```

```
plt.hist(x,bins=40)
plt.title('Histogram')
plt.show()
```

```
plt.hist(y)
plt.title('Histogram')
```

### **Box Plot**

```
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```

### **Pie Chart plot**

```
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0.1, 0)  # explode 1st slice

#plot the pie
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
        autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()


```
